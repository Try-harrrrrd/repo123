```python
ax = sns.boxplot(
    x=df.columns[-1],
    y=df.columns[2],
    data=df
)

sns.stripplot(
    x=df.columns[-1],
    y=df.columns[2],
    data=df,
    jitter=True,
    color='black'
)

plt.show()
```

```python


plt.figure(figsize=(8,6))

species = df.iloc[:,-1].unique()
colors = ['red','green','blue']

for i, sp in enumerate(species):
    temp = df[df.iloc[:,-1] == sp]

    plt.scatter(
        temp.iloc[:,0],
        temp.iloc[:,1],
        color=colors[i],
        label=sp
    )

plt.xlabel(df.columns[0])
plt.ylabel(df.columns[1])
plt.title('Scatter Plot by Species')
plt.legend()

plt.show()
```
