```python
import numpy as np
a = np.array([1, 2, 3, 4, 5])
b = np.array([10, 20, 30, 40, 50])

print("Array a:", a)
print("Array b:", b)
print("\n--- Array Math ---")
print("Addition:       ", a + b)
print("Subtraction:    ", a - b)
print("Multiplication: ", a * b)
print("Division:       ", a / b)
print("Power (a2):   ", a ** 2)
print("Modulo (b % a): ", b % a)
print("Square root (a):", np.sqrt(a))
print("Sum of a:       ", np.sum(a))
print("Mean of b:      ", np.mean(b))
print("Max of a:       ", np.max(a))
print("Min of b:       ", np.min(b))
print("Dot product:    ", np.dot(a, b))
```

```python
# Create two matrices
A = np.array([[1, 2],
              [3, 4]])

B = np.array([[5, 6],
              [7, 8]])

print("Matrix A:")
print(A)

print("\nMatrix B:")
print(B)

# Addition
print("\nA + B:")
print(A + B)

# Subtraction
print("\nA - B:")
print(A - B)

```

```python
# Element-wise multiplication
print("\nA * B (element-wise):")
print(A * B)

# Matrix multiplication
print("\nA @ B (matrix multiplication):")
print(A @ B)

# Division (element-wise)
print("\nA / B:")
print(A / B)

# Transpose
print("\nTranspose of A:")
print(A.T)

# Determinant of A
print("\nDeterminant of A:")
print(np.linalg.det(A))
```
