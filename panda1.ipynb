```python
# Grouping and Aggregating
data = {
    'Department': ['HR', 'IT', 'HR', 'IT'],
    'Salary': [40000, 60000, 45000, 70000]
}

df = pd.DataFrame(data)

print(df.groupby('Department')['Salary'].mean())
```

```python
# Merging and Joining

df1 = pd.DataFrame({
    'ID': [101, 102, 103],
    'Name': ['Ravi', 'Priya', 'Kiran']
})

df2 = pd.DataFrame({
    'ID': [101, 102, 104],
    'Salary': [50000, 65000, 70000]
})

print(pd.merge(df1, df2, on='ID'))
```

```python
df1 = pd.DataFrame({
    'Name': ['Rahul', 'Meera']
})

df2 = pd.DataFrame({
    'Age': [23, 27]
})

print(df1.join(df2))
```
