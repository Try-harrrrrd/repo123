```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv(r"C:\Users\Aiswarya\Documents\socialnetads.csv")
```

```python
print("First 5 Rows of Dataset:")
dataset.head()
```

```python
print(dataset.isnull().sum())
```

```python
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.25,
    random_state=0
)
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()

X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.tree import DecisionTreeClassifier

classifier = DecisionTreeClassifier(
    criterion='entropy',
    random_state=0
)

classifier.fit(X_train, y_train)
```

```python
clf2 = DecisionTreeClassifier(
    max_depth=3,
    min_samples_split=10,
    min_samples_leaf=5,
    random_state=42
)

clf2.fit(X_train, y_train)
```

```python
print(classifier.predict(sc.transform([[30, 87000]])))
```

```python
print(clf2.predict(sc.transform([[51, 87000]])))
```

```python
y_pred = classifier.predict(X_test)
y_pred2 = clf2.predict(X_test)
```

```python
print(
    np.concatenate(
        (
            y_pred.reshape(len(y_pred),1),
            y_test.reshape(len(y_test),1)
        ),
        axis=1
    )
)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score

cm = confusion_matrix(y_test, y_pred)

print("Confusion Matrix:")
print(cm)

accuracy = accuracy_score(y_test, y_pred)

print("\nAccuracy =", accuracy)
```

```python
cm2 = confusion_matrix(y_test, y_pred2)

print("Confusion Matrix:")
print(cm2)

accuracy2 = accuracy_score(y_test, y_pred2)

print("\nAccuracy =", accuracy2)
```

```python
from sklearn.inspection import DecisionBoundaryDisplay

DecisionBoundaryDisplay.from_estimator(
    classifier,
    X_train,
    response_method="predict",
    cmap=plt.cm.Spectral,
    alpha=0.8,
    xlabel=dataset.columns[0],
    ylabel=dataset.columns[1]
)

plt.scatter(
    X_train[:,0],
    X_train[:,1],
    c=y_train,
    s=20,
    edgecolors="k"
)

plt.title('Decision Tree Classifier (Training Set)')

plt.show()
```

```python
DecisionBoundaryDisplay.from_estimator(
    classifier,
    X_test,
    response_method="predict",
    cmap=plt.cm.Spectral,
    alpha=0.8,
    xlabel=dataset.columns[0],
    ylabel=dataset.columns[1]
)

plt.scatter(
    X_test[:,0],
    X_test[:,1],
    c=y_test,
    s=20,
    edgecolors="k"
)

plt.title('Decision Tree Classifier (Testing Set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')

plt.show()
```

```python
from sklearn import tree

plt.figure(figsize=(15,10))

tree.plot_tree(
    classifier,
    filled=True,
    feature_names=['Age','EstimatedSalary'],
    class_names=['Not Purchased','Purchased'],
    rounded=True
)

plt.show()
```

```python
importances = classifier.feature_importances_

print("Feature Importances:")

for name, importance in zip(
    ['Age','EstimatedSalary'],
    importances
):
    print(f"{name}: {importance:.4f}")
```
